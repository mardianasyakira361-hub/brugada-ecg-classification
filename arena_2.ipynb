```python
from sklearn.utils import resample
import pandas as pd
import matplotlib.pyplot as plt

metadata = pd.read_csv("metadata.csv")

healthy = metadata[metadata["brugada"]==0]
brugada = metadata[metadata["brugada"]==1]

brugada_up = resample(brugada, replace=True, n_samples=len(healthy),random_state=42)

metadata_balanced = pd.concat([healthy,brugada_up])

counts = metadata_balanced["brugada"].value_counts()

plt.bar(counts.index,counts.values)

plt.xlabel("class")
plt.ylabel("patients")
plt.title("Distribution Class Brugada")
```

```python
import wfdb
import numpy as np
import pandas as pd
from sklearn.utils import resample

metadata = pd.read_csv("metadata.csv")
healthy = metadata[metadata["brugada"]==0]
brugada = metadata[metadata["brugada"]==1]

brugada_up = resample(brugada, replace=True, n_samples=len(healthy),random_state=42)

metadata_balanced = pd.concat([healthy,brugada_up])

patients = metadata_balanced['patient_id'].tolist()
features_list = []
for pid in patients:
  record = wfdb.rdrecord(f'files/{pid}/{pid}')
  signals = record.p_signal

  feature_dict = {'patient_id':pid}
  for i, lead_name in enumerate(record.sig_name):
    feature_dict[f'{lead_name}_mean'] = np.mean(signals[:,i])
    feature_dict[f'{lead_name}_std'] = np.std(signals[:, i])

  brugada_value = metadata_balanced.loc[metadata_balanced['patient_id']==pid, 'brugada'].values[0]
  feature_dict['brugada']= brugada_value

  features_list.append(feature_dict)

features_df = pd.DataFrame(features_list)

print(features_df.head())
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x = features_df.drop(columns=['patient_id','brugada'])
y = features_df['brugada']

x_train, x_test, y_train, y_test = train_test_split(
    x,y,test_size=0.2,random_state=42, stratify=y
)

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train, y_train)
y_pred = dt_model.predict(x_test)

acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Akurasi: {acc:2f}")
print("confusion Matrix:")
print(cm)
print("Classification Report:")
print(report)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d',cmap='Blues', xticklabels=[0,1], yticklabels=[0,1])
plt.xlabel("predict")
plt.ylabel("actual")
plt.title("Confusion Matrix")
plt.show()

feat_imp = pd.Series(dt_model.feature_importances_,index=x.columns)
feat_imp.sort_values().plot(kind='barh',figsize=(8,6))
plt.title("feature importance")
plt.show()

plt.figure(figsize=(6,4))
plt.hist([y_test, y_pred], label =['actual','prediction'],bins=np.arange(3)-0.5,rwidth=0.6)
plt.xticks([0,1])
plt.xlabel("label")
plt.ylabel("patient")
plt.title("Distribution Actual vs Predict")
plt.legend()
plt.show()
```
